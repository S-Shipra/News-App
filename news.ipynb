```python
import requests
import webbrowser
import pyttsx3
from bs4 import BeautifulSoup

# Initialize Text-to-Speech Engine
engine = pyttsx3.init()
default_rate = 150  
engine.setProperty("rate", default_rate)

# API Configuration
api_key = "ee149e7b-c94c-4e36-aa83-fc7b79ccb4c3"
base_url = "https://content.guardianapis.com/search"

# Available categories
categories = {
    "Politics": "politics",
    "Business and Economy": "business",
    "Technology": "technology",
    "Sports": "sport",
    "Entertainment": "culture",
    "Health": "lifeandstyle/health-and-wellbeing",
    "Science": "science"
}

# Available countries
countries = {
    "India": "India",
    "USA": "USA",
    "UK": "UK",
    "France": "France",
    "World": "World"
}

def fetch_news(category, country):
    """Fetches news articles based on the selected category and country."""
    section = categories.get(category)
    country_query = countries.get(country, "")

    if not section:
        print(f"Invalid category: {category}")
        return []

    url = f"{base_url}?section={section}&q={country_query}&api-key={api_key}&show-fields=headline,trailText,thumbnail"

    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        articles = data.get('response', {}).get("results", [])
        return [
            {
                "title": article["fields"]["headline"],
                "summary": article["fields"].get("trailText", ""),
                "url": article.get("webUrl", "")
            }
            for article in articles
        ]
    else:
        print(f"Error fetching news for {category} in {country}. Status Code: {response.status_code}")
        return []

def read_article(url):
    """Extracts article text and reads it aloud."""
    if not url:
        print("Invalid URL.")
        return

    webbrowser.open(url)
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all("p")
        article_text = " ".join([para.text for para in paragraphs[:5]])

        if article_text.strip():
            print("\nReading Article...\n")

            while True:
                try:
                    speed = int(input("Enter speech speed (default is 150, range 100-250): ").strip())
                    if 100 <= speed <= 250:
                        engine.setProperty("rate", speed)
                        break
                    else:
                        print("Invalid speed. Enter a number between 100 and 250.")
                except ValueError:
                    print("Invalid input. Enter a valid number.")

            engine.say(article_text)
            engine.runAndWait()
        else:
            print("No readable content found.")
    else:
        print("Could not fetch the article.")

# Main Program Execution
print("\nAvailable Countries:", ", ".join(countries.keys()))
country = input("Enter country (India, USA, UK, France, World): ").strip()
if country not in countries:
    print("Invalid country. Please try again.")
else:
    print("\nAvailable Categories:", ", ".join(categories.keys()))
    category = input("Enter news category: ").strip()
    
    if category not in categories:
        print("Invalid category. Please try again.")
    else:
        news_articles = fetch_news(category, country)

        if news_articles:
            print("\nTop Articles:")
            for idx, article in enumerate(news_articles[:3], 1):
                print(f"{idx}. {article['title']} ({article['url']})")

            while True:
                try:
                    choice = int(input("\nEnter the article number to read (or 0 to exit): ")) - 1
                    if choice == -1:
                        print("Exiting program.")
                        break
                    elif 0 <= choice < len(news_articles):
                        read_article(news_articles[choice]["url"])
                        break
                    else:
                        print("Invalid selection. Enter a valid number.")
                except ValueError:
                    print("Invalid input. Enter a number.")
```

```python
import requests
import webbrowser
import pyttsx3
from bs4 import BeautifulSoup
import whisper
from jiwer import wer
import os
import time

# ---------------------------- TTS Engine Setup ---------------------------- #
engine = pyttsx3.init()
engine.setProperty("rate", 150)

# ---------------------------- API Configuration ---------------------------- #
api_key = "ee149e7b-c94c-4e36-aa83-fc7b79ccb4c3"
base_url = "https://content.guardianapis.com/search"

categories = {
    "Politics": "politics",
    "Business and Economy": "business",
    "Technology": "technology",
    "Sports": "sport",
    "Entertainment": "culture",
    "Health": "lifeandstyle/health-and-wellbeing",
    "Science": "science"
}

countries = {
    "India": "India",
    "USA": "USA",
    "UK": "UK",
    "France": "France",
    "World": "World"
}

# ---------------------------- Fetch News ---------------------------- #
def fetch_news(category=None, country=None, search_query=None):
    params = {"api-key": api_key, "show-fields": "headline,trailText,thumbnail"}
    
    if search_query:
        params["q"] = search_query
    elif category and country:
        params["section"] = categories.get(category, "")
        params["q"] = countries.get(country, "")
    
    response = requests.get(base_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        articles = data.get('response', {}).get("results", [])
        return [
            {
                "title": article["fields"]["headline"],
                "summary": article["fields"].get("trailText", ""),
                "url": article.get("webUrl", "")
            }
            for article in articles
        ]
    else:
        print("Error fetching news.")
        return []

# ---------------------------- Transcribe Audio ---------------------------- #
def transcribe_audio(file_path="tts_output.wav"):
    print("ðŸŽ§ Transcribing the saved audio...")
    if not os.path.exists(file_path):
        print("âŒ Audio file not found.")
        return ""
    
    model = whisper.load_model("base")
    result = model.transcribe(file_path)
    
    if not result["text"].strip():
        print("âŒ Whisper couldn't transcribe any text.")
        return ""
    
    print("ðŸ“ Transcribed Text:\n", result["text"])
    return result["text"]

# ---------------------------- Evaluate WER ---------------------------- #
def evaluate_tts(reference_text, hypothesis_text):
    if not reference_text.strip() or not hypothesis_text.strip():
        print("âŒ One of the texts is empty. Cannot compute WER.")
        return
    error = wer(reference_text, hypothesis_text)
    print(f"\nðŸ” Word Error Rate (WER): {error * 100:.2f}%")

# ---------------------------- Read and Evaluate Article ---------------------------- #
def read_article(url):
    if not url:
        print("Invalid URL.")
        return
    
    response = requests.get(url)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, "html.parser")
        paragraphs = soup.find_all("p")
        article_text = " ".join([para.text for para in paragraphs[:5]])

        if not article_text.strip():
            print("âŒ No readable content found.")
            return

        print("\nðŸ“ Extracted Text Preview:\n", article_text[:300])
        
        # Save audio
        print("\nðŸ”Š Saving TTS output...")
        engine.save_to_file(article_text, "tts_output.wav")
        engine.runAndWait()

        # Confirm audio is saved
        time.sleep(2)
        if not os.path.exists("tts_output.wav") or os.path.getsize("tts_output.wav") == 0:
            print("âŒ Audio file not created properly.")
            return
        print("âœ… Audio saved to tts_output.wav")

        # Transcribe and evaluate
        transcript = transcribe_audio("tts_output.wav")
        evaluate_tts(article_text, transcript)
    else:
        print("âŒ Could not fetch the article.")

# ---------------------------- Main Program ---------------------------- #
print("\nAvailable Options:")
print("1. Breaking News")
print("2. Search News")
print("3. Browse by Category & Country")
choice = input("Enter your choice (1/2/3): ").strip()

if choice == "1":
    news_articles = fetch_news(country="World")
elif choice == "2":
    search_query = input("Enter search topic: ").strip()
    news_articles = fetch_news(search_query=search_query)
elif choice == "3":
    print("\nAvailable Countries:", ", ".join(countries.keys()))
    country = input("Enter country: ").strip()
    if country not in countries:
        print("Invalid country. Exiting.")
        exit()
    print("\nAvailable Categories:", ", ".join(categories.keys()))
    category = input("Enter news category: ").strip()
    if category not in categories:
        print("Invalid category. Exiting.")
        exit()
    news_articles = fetch_news(category, country)
else:
    print("Invalid option. Exiting.")
    exit()

# ---------------------------- Article Selection ---------------------------- #
if news_articles:
    print("\nTop Articles:")
    for idx, article in enumerate(news_articles[:3], 1):
        print(f"{idx}. {article['title']} ({article['url']})")
    
    while True:
        try:
            choice = int(input("\nEnter article number to read (or 0 to exit): ")) - 1
            if choice == -1:
                print("Exiting program.")
                break
            elif 0 <= choice < len(news_articles):
                read_article(news_articles[choice]["url"])
                break
            else:
                print("Invalid selection. Try again.")
        except ValueError:
            print("Invalid input. Enter a number.")
else:
    print("No news articles found.")
```
